```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Veri setini oku
data = pd.read_excel('kayseri_kaza_verileri.xlsx')

# Sadece sayısal sütunları seç
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Eksik değerleri ortalama ile doldur
numeric_data.fillna(numeric_data.mean(), inplace=True)

# Kategorik sütunları uygun şekilde kodla (gerekiyorsa)
# Örneğin: data['KAZA TURU'] = data['KAZA TURU'].map({'Yaralamali': 0, 'Ölümlü': 1, 'Hasarli': 2})

# Verileri standartlaştır
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# K-means modelini oluştur ve eğit
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)

# Küme merkezlerini ve etiketleri al
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Küme merkezlerini ve etiketleri yazdır
print("Küme Merkezleri:")
print(cluster_centers)
print("\nEtiketler:")
print(labels)
```
